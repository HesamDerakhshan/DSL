```python
import numpy as np
import pylab as plt
from scipy.stats import mode,skew,kurtosis
```

### Mean, Median, Mode: What They Are ?
* The mean is the average of a data set.
* The mode is the most common number in a data set.
* The median is the middle of the set of numbers.


### Degree 1: The first standardized moment is a measure of mean 
	
${\displaystyle {\tilde {\mu }}_{1}={\frac {\mu _{1}}{\sigma ^{1}}}={\frac {\operatorname {E} \left[(X-\mu )^{1}\right]}{(\operatorname {E} \left[(X-\mu )^{2}\right])^{1/2}}}={\frac {\mu -\mu }{\sqrt {\operatorname {E} \left[(X-\mu )^{2}\right]}}}=0}$	 

### Degree 2: The second standardized moment is is a measure of variance Ïƒ2.
    
${\displaystyle {\tilde {\mu }}_{2}={\frac {\mu _{2}}{\sigma ^{2}}}={\frac {\operatorname {E} \left[(X-\mu )^{2}\right]}{(\operatorname {E} \left[(X-\mu )^{2}\right])^{2/2}}}=1}$	

### Degree 3: The third standardized moment is a measure of skewness.

${\displaystyle {\tilde {\mu }}_{3}={\frac {\mu _{3}}{\sigma ^{3}}}={\frac {\operatorname {E} \left[(X-\mu )^{3}\right]}{(\operatorname {E} \left[(X-\mu )^{2}\right])^{3/2}}}}$	

### Degree 4:	The fourth standardized moment refers to the kurtosis.

${\displaystyle {\tilde {\mu }}_{4}={\frac {\mu _{4}}{\sigma ^{4}}}={\frac {\operatorname {E} \left[(X-\mu )^{4}\right]}{(\operatorname {E} \left[(X-\mu )^{2}\right])^{4/2}}}}$

```python
def ruler1(truth):
    obs = truth+np.random.normal(0,0.03)
    return obs

def ruler2(truth):
    obs = truth+np.random.normal(0.05,0.03)
    return obs

def ruler3(truth):
    obs = truth+np.random.normal(0,0.01)
    return obs

def ruler4(truth):
    obs = truth+np.random.normal(0.015,0.01)
    return obs
# allternative method  
# from rulers import *
```

```python
truth = 10

obs = ruler1(truth)
print(obs)
```

```python
ntry = 100
obss1 = []
obss2 = []
obss3 = []
obss4 = []
for i in range(ntry):
    obss1.append(ruler1(truth))
    obss2.append(ruler2(truth))
    obss3.append(ruler3(truth))
    obss4.append(ruler4(truth))
```

```python
plt.plot([10,10],[0,5],'k--')
plt.plot(obss1,[1]*ntry,'b.')
plt.plot(obss2,[2]*ntry,'r.')
plt.plot(obss3,[3]*ntry,'g.')
plt.plot(obss4,[4]*ntry,'y.')
```
